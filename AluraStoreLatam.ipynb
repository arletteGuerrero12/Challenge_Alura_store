### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

# Calculamos el ingreso total de cada tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostramos los resultados
print(f"Ingreso total de Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total de Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total de Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total de Tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python

productos_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count()
productos_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count()
productos_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count()
productos_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count()


# Mostrar los resultados
print("Categorías y cantidad de productos vendidos en Tienda 1:")
print(productos_categoria_tienda1)

print("\nCategorías y cantidad de productos vendidos en Tienda 2:")
print(productos_categoria_tienda2)

print("\nCategorías y cantidad de productos vendidos en Tienda 3:")
print(productos_categoria_tienda3)

print("\nCategorías y cantidad de productos vendidos en Tienda 4:")
print(productos_categoria_tienda4)
```

```python
# Calcular el total de productos vendidos en cada tienda
total_ventas_tienda = tienda['Cantidad de cuotas'].sum()
total_ventas_tienda2 = tienda2['Cantidad de cuotas'].sum()
total_ventas_tienda3 = tienda3['Cantidad de cuotas'].sum()
total_ventas_tienda4 = tienda4['Cantidad de cuotas'].sum()

# Guardar los resultados en un diccionario para ordenarlos
ventas_totales = {
    "Tienda 1": total_ventas_tienda,
    "Tienda 2": total_ventas_tienda2,
    "Tienda 3": total_ventas_tienda3,
    "Tienda 4": total_ventas_tienda4
}

# Identificar la tienda con más ventas
tienda_mas_vendedora = max(ventas_totales, key=ventas_totales.get)
tienda_menos_vendedora = min(ventas_totales, key=ventas_totales.get)

# Mostrar resultados
print(f"La tienda con más ventas es {tienda_mas_vendedora} con un total de {ventas_totales[tienda_mas_vendedora]} productos vendidos.")
print(f"La tienda con menos ventas es {tienda_menos_vendedora} con un total de {ventas_totales[tienda_menos_vendedora]} productos vendidos.")
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

```python
# Función para encontrar productos más y menos vendidos
def analizar_ventas_por_producto(tienda, nombre_tienda):
    # Agrupa por producto y cuenta las ventas
    ventas_por_producto = tienda.groupby('Producto')['Cantidad de cuotas'].sum()

    # Identifica el producto más y menos vendido
    producto_mas_vendido = ventas_por_producto.idxmax()
    producto_menos_vendido = ventas_por_producto.idxmin()
    cantidad_mas_vendida = ventas_por_producto.max()
    cantidad_menos_vendida = ventas_por_producto.min()

    print(f"\nEn {nombre_tienda}:")
    print(f"Producto más vendido: {producto_mas_vendido} con {cantidad_mas_vendida} ventas.")
    print(f"Producto menos vendido: {producto_menos_vendido} con {cantidad_menos_vendida} ventas.")

    # Visualización de las ventas
    ventas_por_producto.sort_values(ascending=False).plot(kind='bar', figsize=(20, 6))
    plt.title(f"Ventas por Producto en {nombre_tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad Vendida")
    plt.xticks(rotation=45)
    plt.show()

# Analizamos cada tienda
analizar_ventas_por_producto(tienda, "Tienda 1")
analizar_ventas_por_producto(tienda2, "Tienda 2")
analizar_ventas_por_producto(tienda3, "Tienda 3")
analizar_ventas_por_producto(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt

# Cálculo del costo de envío promedio para cada tienda
costos_promedios = {
    "Tienda 1": tienda['Costo de envío'].mean(),
    "Tienda 2": tienda2['Costo de envío'].mean(),
    "Tienda 3": tienda3['Costo de envío'].mean(),
    "Tienda 4": tienda4['Costo de envío'].mean()
}

# Datos para la gráfica
labels = costos_promedios.keys()
values = costos_promedios.values()

# Gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=["gold", "skyblue", "lightcoral", "mediumseagreen"])
plt.title("Porcentaje de Costos de Envío Promedio por Tienda", fontsize=16)
plt.axis("equal")  # Asegura que el gráfico sea circular
plt.show()
```

```python
# Calculamos el costo de envío promedio para cada tienda
def costo_envio_promedio(tienda, nombre_tienda):
    if 'Costo de envío' in tienda.columns:
        costo_promedio = tienda['Costo de envío'].mean()
        print(f"El costo de envío promedio en {nombre_tienda} es: ${round(costo_promedio, 2)}")
    else:
        print(f"No se encontró la columna 'Costo de envío' en {nombre_tienda}.")

# Analizamos cada tienda
costo_envio_promedio(tienda, "Tienda 1")
costo_envio_promedio(tienda2, "Tienda 2")
costo_envio_promedio(tienda3, "Tienda 3")
costo_envio_promedio(tienda4, "Tienda 4")
```
